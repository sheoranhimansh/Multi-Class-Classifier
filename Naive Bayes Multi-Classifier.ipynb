```python
from __future__ import division
import cv2
import os
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
from skimage import color
import matplotlib.pyplot as plt 
from pylab import imread,subplot,imshow,title,gray,figure,show,NullLocator
import scipy.misc
import math
```

```python
def calc_prob(x,mean,stdev):
    xmean = x-mean
    stdev1 = math.pow(stdev,2)
    sigma = 2*stdev1
    xmeansq = math.pow(xmean,2)
    frac = (xmeansq)/(sigma)
    frac = (-1)*frac
    exponent = math.exp(frac)
    constant = math.sqrt(2*math.pi)
    constant = constant*stdev
    ans = 1/constant
    ans = ans*exponent
    return ans
```

```python
input_file = open("sample_train_dataset.txt", "r")
X = []
X_name = []
for file in input_file:
    im = Image.open(file.split()[0]).convert("L").resize((32,32))
    array = np.asarray(im)
    X.append(array.reshape([-1]))
    name = str(file.split()[1])
    X_name.append(name)
    im.close
X = np.array(X)
X_name = np.array(X_name)
```

```python
test_file = open("sample_test_dataset.txt", "r")
Y = []
Y_name = []
for file in test_file:
    im = Image.open(str(file.split('\n')[0])).convert('L')
    array = np.asarray(im)
    Y.append(array.reshape([-1]))
    Y_name.append(str(file.split('\n')[0]))
    im.close
Y_name
```

```python
mean = np.mean(X,axis=0)
mean_x = X - mean
v, eigenvalues ,eigenvectors  = np.linalg.svd(mean_x)
eigenvectors= eigenvectors.T
x_coeff = np.matmul(mean_x,eigenvectors)
eigenvectors = eigenvectors.T
X_new = np.array(x_coeff[:,:32])
eigenvectors = eigenvectors[0:32,:]
```

```python
mean_y = Y - mean
y_coeff = np.matmul(mean_y , eigenvectors.T)
Y_new = y_coeff
```

```python
label_bucket = dict()
for i in range(len(X_new)):
    name = X_name[i] 
    print(name)
    im = X_new[i]
    if name not in label_bucket.keys():
        label_bucket[name] = list()
    label_bucket[name].append(im)

```

```python
mean_bucket = dict()
classes = list()
for i in label_bucket.keys():
    mean_bucket[i] = np.mean(label_bucket[i],axis=0)
    classes.append(i)

classes
```

```python
variance_bucket = dict()
for i in label_bucket.keys():
    variance_bucket[i] = np.std(label_bucket[i],axis=0,ddof=1)
# variance_bucket
```

```python
idx = 0
accuracy = 0
k = 0
for im in Y_new:
    final_prob_arr = []
    for i in label_bucket.keys():
        total_prob = 1
        for j in range(32):
            x = im[j]
            mean = mean_bucket[i][j]
            var = variance_bucket[i][j]
            prob = calc_prob(x,mean,var)
            total_prob *= prob
        num = len(label_bucket[i])
        den = len((X_new))
        prob_class = num/den
        final_prob = total_prob*prob_class
        final_prob_arr.append(final_prob)
    max_ind = max(final_prob_arr)
    ind = final_prob_arr.index(max_ind)
    print(Y_name[k][12])
    if(classes[ind] == Y_name[k][12]):
#         print(Y_name[k][12])
        accuracy += 1
    k += 1
print(accuracy)
```
